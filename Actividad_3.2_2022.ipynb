```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Se saltan las primeras tres filas ya que no tienen datos y se elogio la hoja de 2022.

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name= '2022', skiprows = 3)
df
```

Se eliminaron las siguientes columnas por los siguientes motivos:
Otros no tiene ningun dato 
Folio no es relevante ya que solo se utiliza para rastrear compras internas
UUID es indicador fiscal, pero no es util para el analisis que se desea hacer
TC no es relevante ya que todo esta en MXN
Tipo porque ya hay otra columna que explica el tipo de gasto

```python
df2= df.drop(['Otros', 'UUID', 'TC', 'Tipo', 'Status'], axis=1)
df2
```

```python
valores_nulos=df2.isnull().sum()
valores_nulos
```

Se utilizo forwardfill para rellenar los valores nulos en Poliza, ya que se considero que el primer registro esta lleno y los ultimos 12 no.

```python
df2["Poliza"] =df2["Poliza"].fillna(method="ffill") 
```

```python
valores_nulos=df2.isnull().sum()
valores_nulos
```

Al observar el dataframe, se puede observar que los NAs de MP solo estan cuando en su registro de descripcion es de pago, por lo que se sistituyeron los NAs por el concepto de pago.

```python
df2["MP"] =df2["MP"].fillna("Pago") 
```

```python
valores_nulos=df2.isnull().sum()
valores_nulos
```

Para rellenar los NAs de folio, se puso 99999 para distinguir que aun no se ha asignado un numero de folio.

```python
df2["Folio"] =df2["Folio"].fillna("99999") 
```

```python
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
df2.to_csv("Gastos y costos 2022 sin nulos.csv")
```
